```python
# Reading the contents of the file to understand its structure
with open("tranport_market.SPMapping", 'r') as file:
    content = file.read()

# Displaying the first 1000 characters for a quick glance
content[:1000]
# Parse the XML content
```

```python
import xml.etree.ElementTree as ET
def xml_to_dict_with_attributes(element):
    """Recursively convert an xml element to a dictionary, capturing attributes as well."""
    # Create a dictionary to store the element's data
    data = {}
    
    # Add the element's attributes to the dictionary
    data.update(element.attrib)
    
    # If the element has text content, add it to the dictionary
    if element.text:
        text = element.text.strip()
        if text:
            data['text'] = text

    # If the element has child elements, add them to the dictionary
    for child in element:
        child_data = xml_to_dict_with_attributes(child)
        
        # If this child tag already exists in the dictionary, ensure it's a list and append to it
        if child.tag in data:
            if not isinstance(data[child.tag], list):
                data[child.tag] = [data[child.tag]]
            data[child.tag].append(child_data)
        else:
            data[child.tag] = child_data
    
    return data

root = ET.fromstring(content)
# Convert the XML to dictionary capturing attributes
xml_dict_with_attributes = xml_to_dict_with_attributes(root)
xml_dict_with_attributes
```
